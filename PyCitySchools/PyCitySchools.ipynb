```python
import pandas as pd
from pathlib import Path
```

```python
#import both files
csv_path = Path("..\Starter_Code\PyCitySchools\Resources\schools_complete.csv")
Schools_df = pd.read_csv(csv_path, encoding="UTF-8")

csv_path = Path("..\Starter_Code\PyCitySchools\Resources\students_complete.csv")
students_df = pd.read_csv(csv_path, encoding="UTF-8")
school_data_complete = pd.merge(students_df, Schools_df, how="left", on=["school_name", "school_name"])
school_data_complete=pd.DataFrame(school_data_complete)
school_data_complete
#students_df
```

```python
#Total number of unique schools:
print("Total number schools in the district")
school_data_complete['school_name'].nunique()
```

```python
#Total students
print("Total number of students in the district:")
school_data_complete['size'].sum()
```

```python
#Total budget
print("Total budget of the district:")
school_data_complete['budget'].sum()
```

```python
#Average math score
print("Average Math score of the school district")
school_data_complete['math_score'].mean()
```

```python
#Average reading score
print("Average reading score of the school district")
school_data_complete['reading_score'].mean()
```

```python
#% passing math (the percentage of students who passed math)
pasing_math=school_data_complete.loc[school_data_complete['math_score']>=70,:]

print("Percentage of passing student for math:")

print(str(round( (float(pasing_math['Student ID'].count()) / float(school_data_complete['Student ID'].count()) *100),2))+"%")

#using student ID because two equal grade
```

```python
#% passing reading (the percentage of students who passed reading)
pasing_english=school_data_complete.loc[school_data_complete['reading_score']>=70,:]
print("Percentage of passing student for readin:")
print(str(round( (float(pasing_english['Student ID'].count()) / float(school_data_complete['Student ID'].count()) *100),2))+"%")
```

```python
#% overall passing (the percentage of students who passed math AND reading)
passing_all=school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70),:]
print("Percentage of passing student for all classes:")
print(str(round( (float(passing_all['Student ID'].count()) / float(school_data_complete['Student ID'].count()) *100),2))+"%")
```

## School Summary

```python
#Perform the necessary calculations and then create a DataFrame that summarizes key metrics about each school.
#Include the following:
#School name


group_df = school_data_complete
school_data_complete=school_data_complete.rename(columns={"school_name":"School Name","type":"School Type","size":"Total Students","budget":"Total School Budget"})



group_df=group_df.rename(columns={"school_name":"School Name","type":"School Type","size":"Total Students","budget":"Total School Budget","reading_score":"Average Reading Score","math_score":"Average Math Score"})
group_df=group_df.drop(['School ID','Student ID'], axis=1)

group_df=group_df.reindex(columns=["School Name","School Type","Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score"])


#del group_df['School ID','Student ID']
group_df = group_df.groupby(["School Name"])
group_average=group_df.mean()

per_school_summary=group_average


per_school_summary["Per Student Budget"] = group_average['Total School Budget'] / group_average['Total Students']
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].astype(float).map("${:,.2f}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].astype(float).map("${:,.2f}".format)


pasing_math=school_data_complete.loc[school_data_complete['math_score']>=70, : ]
pasing_reading=school_data_complete.loc[school_data_complete['reading_score']>=70, : ]
pasing_math_count=pasing_math["School Name"].value_counts()
pasing_reading_count=pasing_reading["School Name"].value_counts()


pasing_math_count=pd.DataFrame(pasing_math_count)
per_school_summary["passing math"]=pasing_math_count

pasing_reading_count=pd.DataFrame(pasing_reading_count)
per_school_summary["passing reading"]=pasing_reading_count

pass_all=school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70),:]
pass_all=pass_all["School Name"].value_counts()
pass_all=pd.DataFrame(pass_all)
per_school_summary["Passing Overall"]=pass_all

#school_data_complete = pd.concat([school_data_complete, pasing_math_count], ignore_index=True)
#school_data_complete["School Name"].value_counts()
per_school_summary["Passing Rate Math"]=(per_school_summary["passing math"]/per_school_summary["Total Students"])*100
per_school_summary["Passing Rate Reading"]=(per_school_summary["passing reading"]/per_school_summary["Total Students"])*100
per_school_summary["Passing Rate Overall"]=(per_school_summary["Passing Overall"]/per_school_summary["Total Students"])*100


per_school_summary["Passing Rate Reading"] = per_school_summary["Passing Rate Reading"].astype(float).map("{:,.2f}%".format)
per_school_summary["Passing Rate Math"] = per_school_summary["Passing Rate Math"].astype(float).map("{:,.2f}%".format)
per_school_summary["Passing Rate Overall"] = per_school_summary["Passing Rate Overall"].astype(float).map("{:,.2f}%".format)


#per_school_summary=pasing_math_count
per_school_summary

#school_summary_df
```
